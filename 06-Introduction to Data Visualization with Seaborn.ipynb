<h3>Seaborn <code>relplot()</code></h3>
<ul>
    <li><code>seaborn.relplot()</code> is a versatile function in the Seaborn library used for creating <strong>only relational plots</strong>, which are scatter plots and line plots that show relationships between two variables. </li>
    <li>It is a higher-level function compared to <code>scatterplot()</code> or <code>lineplot()</code> and is often used when you want to create these plots within the framework of Seaborn&apos;s powerful grid system (FacetGrid).</li>
    <li><code>seaborn.relplot()</code> also lets you create <strong>subplots</strong> in a single figure</li>
</ul>
<h4>Key Features of <code>relplot()</code>:</h4>
<ul>
    <li><strong>Supports both scatter and line plots:</strong> You can specify the type of plot you want with the <code>kind</code> parameter (<code>&apos;scatter&apos;</code> or <code>&apos;line&apos;</code>).</li>
    <li><strong>Faceting:</strong> It allows you to create multiple plots divided by the values of another variable (using the <code>col</code> and <code>row</code> parameters).</li>
    <li><strong>Hue, Size, and Style:</strong> You can map variables in the data to different visual aspects of the plot, such as color (<code>hue</code>), size (<code>size</code>), and marker style (<code>style</code>).</li>
</ul>

<h4>Creating subplots in columns using <code>relplot()</code>:</h4>

```python
sns.relplot(
    x="total_bill", 
    y="tip", 
    data=df, 
    kind="scatter",
    col="smoker" #creating subplot based on values of 'smoker' column of dataframe
           )
plt.show()
```

<h4>Creating subplots in rows using <code>relplot()</code>:</h4>

```python
sns.relplot(
    x="total_bill", 
    y="tip", 
    data=df, 
    kind="scatter",
    row="smoker" #creating subplot based on values of 'smoker' column of dataframe
           )
plt.show()
```

<h4>Creating subplots in both rows & columns using <code>relplot()</code>:</h4>

```python
sns.relplot(
    x="total_bill", 
    y="tip", 
    data=df, 
    kind="scatter",
    row="time",
    col="smoker" #creating subplot based on values of 'smoker' column of dataframe
           )
plt.show()
```

<h4>Wrapping columns using <code>col_wrap</code></h4>
<p>Consider the following scenario:</p>
<img src="img/c2.jpeg" width=75%>
<ul>
    <li>Here We have created 4 subplot (column wise) based on the values of <code>df[&quot;day&quot;]</code> which 4 unique values.&nbsp;</li>
    <li>If a column has n number of unique values, then it will create n subplots which looks very congested.</li>
    <li>What if we can do like - In one row there can be maximum two subplot.</li>
    <li>&nbsp;This is possible using <code>col_wrap</code> attribute.</li>
</ul>

```python
sns.relplot(
    x="total_bill", 
    y="tip", 
    data=df, 
    kind="scatter",
    col="day", #creating subplot based on values of 'smoker' column of dataframe
    col_wrap=2,
    height=3 #Height (in inches) of each facet or subplots. 
           )
plt.show()
```

<h4>Subgroups with point size & hue</h4>
<ul>
    <li>We can change size each dot in scatterplot using values of different column. More is size value, the bigger is the point.</li>
    <li>Similarly, we can add <code>hue</code> attribute to add color with different intensity. More is size value, the darker is the point.</li>
    <li>Similarly, we can also provide different dot style using <code>style</code> attribute.</li>
    <li>We can change dot transparency using <code>alpha</code> attribute</li>
</ul>

```python
sns.relplot(
    x="total_bill", 
    y="tip", 
    data=df, 
    kind="scatter",
    size="size", #this will change size each dot based on corresponding value in `size` column of dataframe
    hue="size" #this will add color different intensity according to value in `size` column of dataframe
)
plt.show()
```

<h3>Seaborn <code>catplot()</code></h3>
<ul>
	<li><code>seaborn.catplot()</code> is a powerful and flexible function in the Seaborn library <strong>used to create categorical plots</strong>.&nbsp;</li>
	<li>It is built on top of Seaborn's <strong><code>FacetGrid</code></strong>, which allows for easy creation of multi-plot grids.</li>
    <li>Easily create subplots with <code>col=</code> and <code>row=</code> </li>
</ul>
<h4>Key Features of <code>catplot()</code>:</h4>
<ul>
	<li><strong>Multiple Plot Types</strong>: You can create various categorical plots like bar plots, point plots, box plots, violin plots, and more by specifying the <code>kind</code> parameter.</li>
	<li><strong>Faceting</strong>: You can create grids of plots by faceting on different levels of a categorical variable using the <code>col</code> and <code>row</code> parameters.</li>
	<li><strong>Hue, Size, and Style</strong>: Like other Seaborn functions, <code>catplot()</code> supports the mapping of data to visual properties like color (<code>hue</code>), size (<code>size</code>), and marker style (<code>style</code>).</li>
</ul>
<h4>Types of Plots (<code>kind</code> Parameter):</h4>
<ul>
	<li><strong><code>kind="strip"</code></strong>: Creates a strip plot, which shows all observations along one axis.</li>
	<li><strong><code>kind="swarm"</code></strong>: Creates a swarm plot, which shows all observations along one axis but adjusts the position to avoid overlap.</li>
	<li><strong><code>kind="box"</code></strong>: Creates a box plot, which shows the distribution of data based on quartiles.</li>
	<li><strong><code>kind="violin"</code></strong>: Creates a violin plot, which combines a box plot with a kernel density estimate.</li>
	<li><strong><code>kind="boxen"</code></strong>: Creates a boxen plot, which is optimized for large datasets.</li>
	<li><strong><code>kind="point"</code></strong>: Creates a point plot, which shows point estimates and confidence intervals.</li>
	<li><strong><code>kind="bar"</code></strong>: Creates a bar plot, which shows point estimates and confidence intervals using bars.</li>
	<li><strong><code>kind="count"</code></strong>: Creates a count plot, which shows the count of observations in each categorical bin using bars.</li>
</ul>
<div style="margin-top:30px;">
    <h4>Important Parameters:</h4>
    <ul style="margin-top:5px;">
        <li><strong>data:</strong> The dataset you are using, typically a Pandas DataFrame.</li>
        <li><strong>x, y:</strong> The variables to be plotted on the x and y axes, respectively.</li>
        <li><strong>hue:</strong> Maps different colors to different levels of a categorical variable.</li>
        <li><strong>col, row:</strong> Create subplots in different columns or rows based on the levels of a categorical variable.</li>
        <li><strong>kind:</strong> Specifies the type of plot (as mentioned above).</li>
        <li><strong>height:</strong> Controls the height of each facet.</li>
        <li><strong>aspect:</strong> Controls the aspect ratio of the facets.</li>
        <li><strong>order, hue_order:</strong> Controls the order of categories along the x-axis or for the hue variable.</li>
    </ul>
</div>

<b>Plotting <span style="font-size:20px">Count Plot</span> using <code>seaborn.catplot()</code></b>

```python
sns.catplot(
    x="sex",
    data=df1,
    kind="count"
)
plt.show()
```

```python
# same as above code, but specifying count bar order
sns.catplot(
    x="sex",
    data=df1,
    kind="count",
    order = ["female", "male"]
)
plt.show()
```

						<h3>1. <code>sns.set_style()</code></h3>
						<p>The <code>sns.set_style()</code> function is used to set the overall style of the plots, such as the background, gridlines, and axes. 							<li><strong><code>"darkgrid"</code></strong>: A dark background with gridlines (good for emphasizing data points against a darker background).</li>

<h3>2. <code>sns.set_palette()</code></h3>
<p>The <code>sns.set_palette()</code> function allows you to set the color palette for your plots. This function is particularly useful when you want to ensure that your plots use consistent and visually appealing colors, especially when plotting multiple categories.</p>
<h4><strong>Common Palettes:</strong></h4>
<ul>
	<li><strong>Categorical Palettes</strong>: <code>"deep"</code>, <code>"muted"</code>, <code>"bright"</code>, <code>"pastel"</code>, <code>"dark"</code>, <code>"colorblind"</code>.</li>

	<li><strong>Diverging Palettes</strong>: <code>"RdBu"</code>, <code>"Spectral"</code>, <code>"coolwarm"</code>, etc.</li>
</ul>
<h4><strong>Custom Palettes:</strong></h4>
<p>You can create custom palettes using <code>sns.color_palette()</code> with a list of colors or by specifying a color map.</p>
<h4><strong>Example Usage:</strong></h4>

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set a palette
sns.set_palette("pastel")

# Create a plot using the custom palette
sns.barplot(x=["A", "B", "C"], y=[1, 2, 3])

plt.show()
```

<p>FacetGrid vs. AxesSubplot objects</p>
<p>In Matplotlib and Seaborn, <code>FacetGrid</code> and <code>AxesSubplot</code> are two different types of objects that are used to manage and control the layout of plots, but they serve different purposes and offer distinct capabilities.</p>
<h3>1. <strong>FacetGrid Object</strong></h3>
<ul>
	<li><code>FacetGrid</code> is a Seaborn object designed to manage the creation of complex, multi-plot grids, where each subplot represents a subset of the data.</li>
	<li>It is particularly useful when you want to create the same type of plot for different subsets of your data based on one or more categorical variables.</li>
</ul>
<h3>2. <strong>AxesSubplot Object</strong></h3>
<ul>
	<li><code>AxesSubplot</code> is a Matplotlib object that represents a single plot or subplot within a figure.</li>
	<li>It is the fundamental object for plotting in Matplotlib and is more general-purpose compared to <code>FacetGrid</code>.</li>
</ul>
