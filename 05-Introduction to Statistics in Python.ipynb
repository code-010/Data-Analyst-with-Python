<h2>Types of Statistics</h2>

    <li>Examples include hypothesis testing, confidence intervals, and regression analysis.</li>
</ul>

<h2>Which measure to use? Mean or Median or Mode ?</h2>
<ul>
    <li>Mean is the most frequently used measure of central tendency and generally considered the best measure of it. However, there are some situations where either median or mode are preferred.</li>
    <li>Median is the preferred measure of central tendency when:
        <ul>
            <li>There are a few extreme scores in the distribution of the data. (NOTE: Remember that a single outlier can have a great effect on the mean).</li>
            <li>Another time when we usually prefer the median over the mean (or mode) is when our data is skewed. As the data becomes skewed the mean loses its ability to provide the best central location for the data because the skewed data is dragging it away from the typical value. However, the median best retains this position and is not as strongly influenced by the skewed values.</li>
            <li>There are some missing or undetermined values in your data.</li>
            <li>There is an open-ended distribution (For example, if you have a data field which measures number of children and your options are 00, 11, 22, 33, 44, 55 or “66 or more,” than the “66 or more field” is open ended and makes calculating the mean impossible, since we do not know exact values for this field).</li>
            <li>You have data measured on an ordinal scale</li>
        </ul>
    </li>
    <li>Mode is the preferred measure when data are measured in a nominal ( and even sometimes ordinal) scale.</li>
</ul>

<h2>Skewness</h2>
<ul>
    <li>Skewness is a measure of asymmetry of a distribution that indicates whether the observations in a dataset are concentrated on one side.</li>
    <li><img src="img/pc19.png" width=75%></li>
    <li>A distribution can have right (or positive), left (or negative), or zero skewness. A right-skewed distribution is longer on the right side of its peak, and a left-skewed distribution is longer on the left side of its peak.</li>
    <li>Right (positive) skewness means that the outliers are to the right (long tail to the right).</li>
    <li>Left (negative) skewness means that the outliers are to the left.</li>
    <li>Zero skew means the distribution is symmetrical. Its left and right sides are mirror images.</li>
    <li>How to calculate skewness? 
        <ul>
            <li>There are several formulas to measure skewness. One of the simplest is <b>Pearson’s median skewness</b>. It takes advantage of the fact that the mean and median are unequal in a skewed distribution.</li>
                <div>
                $$
  \text{Pearson’s median skewness} = 3 * \frac{\text{Mean}-\text{Median}}{Standard Deviation}
  $$
                    </div>
            <li>If Pearson’s median <b>skewness</b> == 0, Then <b>Zero Skew</b></li>
            <li>If Pearson’s median <b>skewness</b> > 0, Then <b>Positive Skew</b></li>
            <li>If Pearson’s median <b>skewness</b> < 0, Then <b>Negative Skew</b></li>
            <li>Real observations rarely have a Pearson’s median skewness of exactly 0. If your data has a value close to 0, you can consider it to have zero skew.</li>
        </ul>
    </li>
</ul>

<h2>Probability Distributions in Python</h2>

<h3>1. Uniform Distribution</h3>

<p><strong>Definition:</strong> The uniform distribution is a continuous probability distribution where all outcomes are equally likely within a given range.</p>

<p><strong>Properties:</strong></p>
<ul>
  <li><strong>Probability Density Function (PDF):</strong> \( f(x) = \frac{1}{b-a} \) for \( a \leq x \leq b \)</li>
  <li><strong>Mean:</strong> \( \frac{a+b}{2} \)</li>
  <li><strong>Variance:</strong> \( \frac{(b-a)^2}{12} \)</li>
</ul>

<p><strong>Python Example:</strong></p>

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
a, b = 0, 1  # Range

# Generate random numbers
data = np.random.uniform(a, b, 1000)

# Plot
plt.hist(data, bins=30, density=True, alpha=0.7, color='skyblue')
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

<h3>2. Normal Distribution</h3>

<p><strong>Definition:</strong> The normal distribution, also known as the Gaussian distribution, is a continuous distribution characterized by its bell-shaped curve.</p>

<p><strong>Properties:</strong></p>
<ul>
  <li><strong>Probability Density Function (PDF):</strong> \( f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2} \)</li>
  <li><strong>Mean:</strong> \( \mu \)</li>
  <li><strong>Variance:</strong> \( \sigma^2 \)</li>
</ul>

<p><strong>Python Example:</strong></p>

```python
from scipy.stats import norm

# Parameters
mu, sigma = 0, 1  # Mean and standard deviation

# Generate random numbers
data = np.random.normal(mu, sigma, 1000)

# Plot
x = np.linspace(-4, 4, 100)
pdf = norm.pdf(x, mu, sigma)

plt.hist(data, bins=30, density=True, alpha=0.7, color='lightcoral')
plt.plot(x, pdf, 'k', linewidth=2)
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

<h3>3. Binomial Distribution</h3>

<p><strong>Definition:</strong> The binomial distribution is a discrete distribution that models the number of successes in a fixed number of independent Bernoulli trials.</p>

<p><strong>Properties:</strong></p>
<ul>
  <li><strong>Probability Mass Function (PMF):</strong> \( P(X = k) = \binom{n}{k} p^k (1-p)^{n-k} \)</li>
  <li><strong>Mean:</strong> \( np \)</li>
  <li><strong>Variance:</strong> \( np(1-p) \)</li>
</ul>

<p><strong>Python Example:</strong></p>

```python
from scipy.stats import binom

# Parameters
n, p = 10, 0.5  # Number of trials, probability of success

# Generate random numbers
data = np.random.binomial(n, p, 1000)

# Plot
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)

plt.hist(data, bins=np.arange(-0.5, n+1.5), density=True, alpha=0.7, color='lightgreen', rwidth=0.8)
plt.plot(x, pmf, 'o', color='black')
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.show()
```

<h3>4. Poisson Distribution</h3>

<p><strong>Definition:</strong> The Poisson distribution is a discrete distribution that models the number of events occurring within a fixed interval of time or space.</p>

<p><strong>Properties:</strong></p>
<ul>
  <li><strong>Probability Mass Function (PMF):</strong> \( P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!} \)</li>
  <li><strong>Mean:</strong> \( \lambda \)</li>
  <li><strong>Variance:</strong> \( \lambda \)</li>
</ul>

<p><strong>Python Example:</strong></p>

```python
from scipy.stats import poisson

# Parameter
lambda_ = 3  # Average number of events

# Generate random numbers
data = np.random.poisson(lambda_, 1000)

# Plot
x = np.arange(0, np.max(data)+1)
pmf = poisson.pmf(x, lambda_)

plt.hist(data, bins=np.arange(-0.5, np.max(data)+1.5), density=True, alpha=0.7, color='lightcoral', rwidth=0.8)
plt.plot(x, pmf, 'o', color='black')
plt.title('Poisson Distribution')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.show()
```

<h3>5. Exponential Distribution</h3>

<p><strong>Definition:</strong> The exponential distribution is a continuous distribution that models the time between events in a Poisson process.</p>

<p><strong>Properties:</strong></p>
<ul>
  <li><strong>Probability Density Function (PDF):</strong> \( f(x) = \lambda e^{-\lambda x} \) for \( x \geq 0 \)</li>
  <li><strong>Mean:</strong> \( \frac{1}{\lambda} \)</li>
  <li><strong>Variance:</strong> \( \frac{1}{\lambda^2} \)</li>
</ul>

<p><strong>Python Example:</strong></p>

```python
from scipy.stats import expon

# Parameter
lambda_ = 1  # Rate parameter

# Generate random numbers
data = np.random.exponential(1/lambda_, 1000)

# Plot
x = np.linspace(0, np.max(data), 100)
pdf = expon.pdf(x, scale=1/lambda_)

plt.hist(data, bins=30, density=True, alpha=0.7, color='lightsalmon')
plt.plot(x, pdf, 'k', linewidth=2)
plt.title('Exponential Distribution')
plt.xlabel('Time Between Events')
plt.ylabel('Density')
plt.show()
```

<h2>Understanding PDF and CDF</h2>

<p>Understanding when to use Probability Density Functions (PDFs) and Cumulative Distribution Functions (CDFs) is crucial for analyzing probability distributions. Here’s a breakdown of when to use each and what they represent:</p>

<h3>Probability Density Function (PDF)</h3>

<h4>Definition</h4>
<p>The Probability Density Function (PDF) describes the likelihood of a random variable to take on a specific value. It is used for continuous random variables.</p>

<h4>Characteristics</h4>
<ul>
  <li>The PDF is non-negative for all values.</li>
  <li>The area under the PDF curve over the entire space equals 1, representing the total probability.</li>
  <li>The value of the PDF at a specific point is not a probability, but a density.</li>
  <li>For continuous distributions, the probability of the random variable being exactly any specific value is zero.</li>
</ul>

<h4>When to Use</h4>
<ul>
  <li><strong>Finding Densities:</strong> Use the PDF when you are interested in understanding the relative likelihood of different outcomes, but not the actual probabilities.</li>
  <li><strong>Modeling Distributions:</strong> Use the PDF to define the distribution of continuous random variables and understand the shape (e.g., normal, exponential).</li>
  <li><strong>Probability Calculations:</strong> Calculate probabilities over an interval \([a, b]\) using the integral of the PDF:
    <br><br>
    \[
    P(a \leq X \leq b) = \int_{a}^{b} f(x) \, dx
    \]
  </li>
</ul>

<h4>Example: Normal Distribution</h4>

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the normal distribution
mu, sigma = 0, 1

# Generate values
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, mu, sigma)

# Plot PDF
plt.plot(x, pdf, label='PDF', color='blue')
plt.title('Probability Density Function')
plt.xlabel('x')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

<h3>Cumulative Distribution Function (CDF)</h3>

<h4>Definition</h4>
<p>The Cumulative Distribution Function (CDF) describes the probability that a random variable \(X\) will take a value less than or equal to \(x\).</p>

<h4>Characteristics</h4>
<ul>
  <li>The CDF ranges from 0 to 1.</li>
  <li>It is a non-decreasing function.</li>
  <li>The CDF provides actual probabilities of the random variable being less than or equal to a given value.</li>
</ul>

<h4>When to Use</h4>
<ul>
  <li><strong>Probability Calculations:</strong> Use the CDF to directly find the probability that a random variable is less than or equal to a certain value, \( P(X \leq x) \).</li>
  <li><strong>Percentiles and Quantiles:</strong> Determine percentiles or quantiles, which are specific points in the distribution.</li>
  <li><strong>Comparative Analysis:</strong> Compare two distributions to understand differences in their probability structures.</li>
</ul>

<h4>Example: Normal Distribution</h4>

```python
# Generate values
cdf = norm.cdf(x, mu, sigma)

# Plot CDF
plt.plot(x, cdf, label='CDF', color='red')
plt.title('Cumulative Distribution Function')
plt.xlabel('x')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

<h3>Key Differences</h3>

<ul>
  <li><strong>PDF vs. CDF:</strong>
    <ul>
      <li><strong>PDF</strong> provides the density of the distribution, showing where values are more or less likely to occur.</li>
      <li><strong>CDF</strong> provides cumulative probabilities, useful for calculating probabilities over intervals.</li>
    </ul>
  </li>
  <li><strong>Use Cases:</strong>
    <ul>
      <li><strong>PDF</strong> is often used for visualizing the shape and spread of the distribution.</li>
      <li><strong>CDF</strong> is useful for direct probability calculations and determining percentiles.</li>
    </ul>
  </li>
</ul>

<h3>Summary</h3>
<ul>
  <li>Use <strong>PDF</strong> to analyze the density and shape of continuous probability distributions.</li>
  <li>Use <strong>CDF</strong> when you need cumulative probabilities or when working with quantiles and percentiles.</li>
</ul>

<p>These functions are fundamental for statistical analysis, providing insights into the behavior and properties of random variables.</p>


<h3>Percent Point Function (PPF)</h3>

<ul>
    <li><strong>Definition: </strong>The PPF is the inverse of the CDF. It is used to determine the value of the random variable that corresponds to a given cumulative probability.</li>
    <li><strong>When To Use: </strong>
        <ul>
            <li>To find the value below which a given percentage of observations in a distribution fall.</li>
            <li>Use when you have a probability and want to find the corresponding value in the distribution (e.g., finding percentiles or critical values).</li>
        </ul>
    </li>
    <li><strong>Example: </strong>Determine the 95th percentile of the distribution.</li>
    <li><strong>Example: </strong></li>
</ul>

```python
# Calculate the PPF for a given probability
ppf_value = norm.ppf(0.95, mu, sigma)
print(f"95th percentile: {ppf_value}")
```

<h2>Using rvs() to Generate Random Numbers</h2>

<p>The rvs() function is part of distribution objects in the scipy.stats module. the rvs() function is a method provided by the scipy.stats module in Python to generate random variables following a specified probability distribution. It stands for "random variates" and can be used for both continuous and discrete distributions. Here are examples using various distributions:
</p>
<h3>1. Normal Distribution</h3>
<p>The normal distribution is a continuous probability distribution with a bell-shaped curve.</p>

```python
from scipy.stats import norm
import matplotlib.pyplot as plt

# Parameters
mu, sigma = 0, 1  # Mean and standard deviation

# Generate random numbers
normal_data = norm.rvs(loc=mu, scale=sigma, size=1000)

# Plot
plt.hist(normal_data, bins=30, density=True, alpha=0.7, color='lightcoral')
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

<h3>2. Uniform Distribution</h3>
<p>The uniform distribution is a continuous distribution where all outcomes are equally likely within a given range.</p>

```python
from scipy.stats import uniform

# Parameters
low, high = 0, 1  # Range

# Generate random numbers
uniform_data = uniform.rvs(loc=low, scale=high-low, size=1000)

# Plot
plt.hist(uniform_data, bins=30, density=True, alpha=0.7, color='skyblue')
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

<h3>3. Binomial Distribution</h3>
<p>The binomial distribution models the number of successes in a fixed number of independent Bernoulli trials.</p>

```python
from scipy.stats import binom

# Parameters
n, p = 10, 0.5  # Number of trials, probability of success

# Generate random numbers
binomial_data = binom.rvs(n, p, size=1000)

# Plot
plt.hist(binomial_data, bins=np.arange(-0.5, n+1.5), density=True, alpha=0.7, color='lightgreen', rwidth=0.8)
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.show()
```

<h3>4. Poisson Distribution</h3>
<p>The Poisson distribution models the number of events occurring within a fixed interval.</p>

```python
from scipy.stats import poisson

# Parameter
lambda_ = 3  # Average number of events

# Generate random numbers
poisson_data = poisson.rvs(lambda_, size=1000)

# Plot
plt.hist(poisson_data, bins=np.arange(-0.5, np.max(poisson_data)+1.5), density=True, alpha=0.7, color='lightcoral', rwidth=0.8)
plt.title('Poisson Distribution')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.show()
```

<h3>5. Exponential Distribution</h3>
<p>The exponential distribution models the time between events in a Poisson process.</p>

```python
from scipy.stats import expon

# Parameter
lambda_ = 1  # Rate parameter

# Generate random numbers
exponential_data = expon.rvs(scale=1/lambda_, size=1000)

# Plot
plt.hist(exponential_data, bins=30, density=True, alpha=0.7, color='lightsalmon')
plt.title('Exponential Distribution')
plt.xlabel('Time Between Events')
plt.ylabel('Density')
plt.show()
```
