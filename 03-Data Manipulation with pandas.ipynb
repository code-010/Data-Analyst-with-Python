## Subsetting
1. **By single columns:** Use <code>df[column_name]</code> to select a single column.
2. **By multiple columns:** Use <code>df[[col1, col2]]</code> to select multiple columns.
3. **By rows:** Use slicing with .iloc or .loc to select rows. 
    <ul style="margin-top:0">
    <li>For example, <code>df.iloc[0:5] selects the first 5 rows.</code></li>
    </ul>
4. **Based on text data:** Use boolean indexing with conditions.
    <ul style="margin-top:0">
    <li>For example, <code>df[df['column_name'] == 'value']</code>.</li>
    </ul>
5. **Based on multiple conditions:** Combine conditions using & (and) or | (or). 
    <ul style="margin-top:0">
    <li>For example, <code>df[(df['col1'] > 10) & (df['col2'] < 5)]</code>.</li>
    </ul>
6. **Using .isin():** Check if values in a column are in a list. 
    <ul style="margin-top:0">
    <li>For example, <code>df[df['column_name'].isin([value1, value2])]</code>.</li>
    </ul>

## Summarizing Numerical Data
<ul>
    <li>
        <b>Using .median(), .mode(), .min(), .max(), .var(), .std(), .sum(), .quantile()</b>
        <ul>
            <li><code>.median()</code>: Returns the median of the data.</li>
            <li><code>.mode()</code>: Returns the mode(s) of the data.</li>
            <li><code>.min()</code>: Returns the minimum value.</li>
            <li><code>.max()</code>: Returns the maximum value.</li>
            <li><code>.var()</code>: Returns the variance.</li>
            <li><code>.std()</code>: Returns the standard deviation.</li>
            <li><code>.sum()</code>: Returns the sum of the values.</li>
            <li><code>.quantile()</code>: Returns the value at a given quantile.</li>
        </ul>
    </li>
    <li>
        <b>The .agg() Method</b>: Allows one or more aggregation operations on the DataFrame.
    </li>
    <li>
        <b>Summaries on Multiple Columns</b>: Use the <code>.agg()</code> method to summarize multiple columns.
    </li>
    <li>
        <b>Multiple Summaries</b>: Combine multiple summary statistics in one operation using the <code>.agg()</code> method.
    </li>
</ul>



```python
#Load Data
loan_df = pd.read_csv("data/loan_data.csv")
loan_df.head()
```

```python
print(f"Median : {loan_df['installment'].median()}")
print(f"Mode : {loan_df['installment'].mode()}") #Returns as Series
print(f"Min : {loan_df['installment'].min()}")
print(f"Max : {loan_df['installment'].max()}")
print(f"Varience : {loan_df['installment'].var()}")
print(f"Standard Deviation : {loan_df['installment'].std()}")
print(f"Total Sum : {loan_df['installment'].sum()}")
print(f"Quantile : {loan_df['installment'].quantile(q=0.25)}") #default q=0.50
```

```python
#The .agg() Method
print(loan_df["installment"].agg([np.median, np.max]))

#We can also provide user-defined aggregation function through .agg()
def my_avg(column):
    temp=0
    for i in column:
        temp += i
    return temp/len(column)

print(f'Custom Aggregation : {loan_df["installment"].agg(my_avg)}')
```

```python
#Summaries on Multiple Columns
loan_df[["installment", "fico"]].agg(np.mean)
```

```python
#Multiple Summaries
print(loan_df[["installment", "fico"]].agg([np.median, my_avg]))
```

```python
df.pivot_table(values="Budget", index="Genre") #Default aggfunc=mean
```

```python
#Single Statistics in aggfunc
df.pivot_table(values="Budget", index="Genre", aggfunc=np.median)
```

```python
#Multiple Statistics in aggfunc
df.pivot_table(values="Budget", index="Genre", aggfunc=[np.median, np.mean])
```

```python
#Pivot on Two Variables

#Following code is same as df.groupby(by=["Genre","Lead Studio"])["Budget"].agg(np.median)
df.pivot_table(values="Budget", index="Genre", columns="Lead Studio", aggfunc=np.median)
```

```python
#Filling Missing Values in Pivot Tables
df.pivot_table(values="Budget", index="Genre", columns="Lead Studio", aggfunc=np.median, fill_value=0)
```

```python
#Summing with Pivot Tables
df.pivot_table(values="Budget", index="Genre", columns="Lead Studio", aggfunc=np.mean, fill_value=0, margins=True)
```

## Multi-level Indexes
<ul>
    <li>
        <b>Subset the outer level with a list: </b>
        Use <code>df.loc[[outer_level_value_1, outer_level_value_2, ...]]</code> to subset.
    </li>
    <li>
        <b>Subset inner levels with a list of tuples: </b>
        Use <code>df.loc[[(outer_level_value, inner_level_value_1), (outer_level_value, inner_level_value_2), ...]]</code> to subset.
    </li>
    <li>
        <b>Sorting by index values: </b>
        Use <code>sort_index()</code> to sort DataFrame by index values.
    </li>
    <li>
        <b>Controlling <code>sort_index</code>: </b>
        Use parameters like <code>level</code> and <code>ascending</code> in <code>sort_index()</code> to contrul sorting behavior.
    </li>
</ul>

```python
# Creating Multi-level Index temperatures by country & city
temperatures_ind = temperatures.set_index(["country", "city"])
temperatures_ind
```

```python
#Subset the outer level with a list
temperatures_ind.loc[["China", "India"]]
```

```python
#Subset inner levels with a list of tuples
temperatures_ind.loc[[("Brazil", "Rio De Janeiro"), ("India", "Surat")]]
```

```python
#Sorting by index values

# Sort temperatures_ind by index values
temperatures_srt = temperatures_ind.sort_index()
temperatures_srt
```

```python
# Sort temperatures_ind by index values at the city level
temperatures_ind.sort_index(level="city") #level value can be specified as Integer
```

```python
#Controlling sort_index

# Sort temperatures_ind by country then descending city
temperatures_ind.sort_index(level=["country", "city"], ascending=[True, False])
```

## Slicing and Subsetting with <code>.loc</code> and <code>.iloc</code>
<ul>
    <li>
        <b>Slicing for single level index: </b>
        <ul type="i">
            <li>
                <h4 style="margin-top:0">Subsetting by row/culumn number</h4>
                Use <code>df.iloc[row_start:row_end, cul_start:cul_end]</code> for subsetting.
            </li>
            <li>
                <h4 style="margin-top:0"><code>.loc[]</code> + slicing</h4>
                Use <code>df.loc[row_start:row_end, cul_start:cul_end]</code> for subsetting.
            </li>
        </ul>
    </li>
    <li>
        <b>Slicing the outer index level for Multi-level indexes: </b>
        Use <code>df.loc[(outer_level_start:outer_level_end)]</code> for slicing.
    </li>
    <li>
        <b>Slicing the inner index levels for Multi-level indexes: </b>
        Use <code>pd.IndexSlice</code> or <code>slice(None)</code> to slice inner levels.
    </li>
    <li>
        <b>Slicing culumns: </b>
        Use <code>df.loc[:, cul_start:cul_end]</code> or <code>df.iloc[:, cul_start:cul_end]</code> to slice culumns.
    </li>
    <li>
        <b>Slicing by dates: </b>
        <ul type="i">
            <li>
                <h4 style="margin-top:0">Slicing by partial dates</h4>
                Use <code>df.loc['YYYY']</code>, <code>df.loc['YYYY-MM']</code>, or <code>df.loc['YYYY-MM-DD']</code> for partial date slicing.
            </li>
        </ul>
    </li>
</ul>

```python
#Slicing for single level index:
#Subsetting by row/culumn number

temperatures.loc[1:5, ["date","city","avg_temp_c"]]
```

```python
#Subsetting by row/culumn number
temperatures.loc[[1, 100], ["date","city","avg_temp_c"]]
```

```python
#Load data
avocados = pd.read_pickle("data/avoplotto.pkl")
avocados.head()
```

```python
# Histogram of conventional avg_price 
avocados[avocados.type=="conventional"]["avg_price"].hist()

# Histogram of organic avg_price
avocados[avocados.type=="organic"]["avg_price"].hist()

# Add a legend
plt.legend(["conventional", "organic"])

# Show the plot
plt.show()
```

```python
# Get the total number of avocados sold of each size
nb_sold_by_size = avocados.groupby("size")["nb_sold"].sum()

# Create a bar plot of the number of avocados sold by size
nb_sold_by_size.plot(kind="bar")
```

```python
# Get the total number of avocados sold on each date
nb_sold_by_date = avocados.groupby("date")["nb_sold"].sum()

# Create a line plot of the number of avocados sold by date
nb_sold_by_date.plot(kind="line", rot=45)

# Show the plot
plt.show()
```

## Missing values in pandas DataFrames
<ul>
    <li>
        <b>Detecting missing values: </b>
        Use <code>df.isnull()</code> to detect missing values in the DataFrame. It returns a DataFrame of the same shape with boolean values indicating the presence of missing values.
    </li>
    <li>
        <b>Detecting any missing values: </b>
        Use <code>df.isnull().any()</code> to check if there are any missing values in the DataFrame. It returns a boolean value for each column.
    </li>
    <li>
        <b>Counting missing values: </b>
        Use <code>df.isnull().sum()</code> to count the number of missing values in each column of the DataFrame.
    </li>
</ul>

```python
# Check individual values for missing values
avocados.isna()
```

```python
# Check each column for missing values
avocados.isna().any()
```

```python
# Count of missing values by variable
avocados.isna().sum()
```

## Handling Missing values
<ul>
    <li>
        <b>Removing missing values: </b>
        Use <code>df.dropna()</code> to remove rows with missing values. Use <code>df.dropna(axis=1)</code> to remove columns with missing values.
    </li>
    <li>
        <b>Replacing missing values: </b>
        Use <code>df.fillna(value)</code> to replace missing values with a specified value. You can also use <code>df.fillna(method='ffill')</code> to forward fill or <code>df.fillna(method='bfill')</code> to backward fill missing values.
    </li>
</ul>
